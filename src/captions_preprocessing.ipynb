# Captions preprocessing

KauiRec dataset gives us the ability to manipulate captions and title covers of all the videos. However, captions contain a lot of weird characters (such as emoji) and it is written in simplified chinese!

So when I started to make my loader functions to load the dataset in my python scripts, I had issues using Pandas on the `luairec_caption_category.csv` file because of issues with encoding. I had to make little tweaks so that I can open this file and manipulate this data.

The aim here is to first open and read the CSV file, then remove the emojis and finally save the file again with the right encoding method. This notebook also come with few other preprocessing line codes that will be applied when loading the dataset in `dataset_load.py` python script.

```python
import pandas as pd
# Emoji library that comes with great functions to manipulate emojis very easily
import emoji
```

```python
# Loading the captions dataset file
with open("../kuairec/data/kuairec_caption_category.csv", mode='r') as f:
    lines = f.readlines()
    file_content = ''.join(lines)
    # Remove emojis and remove also cells equal to 'UNKNOWN' indecating that we do not have any value here
    file_content = emoji.replace_emoji(file_content.replace('UNKNOWN', ''), '')
```

```python
# Save file to a new one that will be readable by Pandas
with open("../kuairec/data/caption_category_readable.csv", mode='w') as f:
    f.write(file_content)
```

```python
df = pd.read_csv('../kuairec/data/caption_category_readable.csv')
```

```python
df['video_id'] = df.index
```

```python
# Fill nan values with negative values so that we can filter them easily later
df['first_level_category_id'] = df['first_level_category_id'].fillna(-1).astype(int)
df['second_level_category_id'] = df['second_level_category_id'].fillna(-1).astype(int)
df['third_level_category_id'] = df['third_level_category_id'].fillna(-1).astype(int)
```

```python
# Remove all the ASCII characters because sometimes captions contained a mix of chinese and ASCII chars
df['manual_cover_text'] = df['manual_cover_text'].str.replace(r'[\x00-\x7F]', '', regex=True)
df['caption'] = df['caption'].str.replace(r'[\x00-\x7F]', '', regex=True)
df['topic_tag'] = df['topic_tag'].str.replace(r'[\x00-\x7F]', '', regex=True)
df['first_level_category_name'] = df['first_level_category_name'].str.replace(r'[\x00-\x7F]', '', regex=True)
df['second_level_category_name'] = df['second_level_category_name'].str.replace(r'[\x00-\x7F]', '', regex=True)
df['third_level_category_name'] = df['third_level_category_name'].str.replace(r'[\x00-\x7F]', '', regex=True)
```
